# UNIVERSITY OF PORTSMOUTH

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
from selenium.common.exceptions import TimeoutException

# Set up Chrome webdriver
driver = webdriver.Chrome()

# Starting URL
base_url = "https://www.port.ac.uk/study/courses?page="
page_number = 1

# Initialize lists to store course details
course_names = []
awards = []
start_dates = []
course_links=[]

while True:
    # Construct URL for each page
    url = base_url + str(page_number)
    driver.get(url)

    try:
        # Find all course containers
        wait = WebDriverWait(driver, 10)
        course_containers = wait.until(EC.visibility_of_all_elements_located((By.CLASS_NAME, "ais-InfiniteHits-item")))
    except TimeoutException:
        break

    # Extract course details and store them in lists
    for course in course_containers:
        # Course name
        course_name_element = course.find_element(By.CLASS_NAME, "card__heading--main")
        course_name = course_name_element.text.strip()
        course_names.append(course_name)


        # Award
        try:
            award_element = course.find_element(By.XPATH, ".//p[contains(., 'Award:')]")
            award = award_element.text.split(":")[1].strip()
        except:
            award = "NA"
        awards.append(award)

        # Start date
        try:
            start_date_element = course.find_element(By.XPATH, ".//p[contains(., 'Start date:')]/span[@class='card__start-date--month']")
            start_date = start_date_element.text.strip()
        except:
            start_date = "NA"
        start_dates.append(start_date)
        
        try:
            
            course_link_element = course.find_element(By.CSS_SELECTOR,'a.button--overlay')
            course_link = course_link_element.get_attribute('href')
        except:
            course_link="NA"
        course_links.append(course_link)

    # Move to the next page
    page_number += 1

# Close the webdriver
driver.quit()

# Create a dataframe
data = {
    "Course_Title": course_names,
    "Degree_Type": awards,
    "Start_Date": start_dates,
    "Course_Link": course_links
}

df = pd.DataFrame(data)

# Print the dataframe
df
```

```python
# Save the data to a CSV file
df.to_csv("UniversityOfPortsmouth.csv", index=False)
```
