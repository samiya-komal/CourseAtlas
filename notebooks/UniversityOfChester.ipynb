# University Of Chester

```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from datetime import datetime

driver = webdriver.Chrome()
base_url = 'https://www1.chester.ac.uk/course_atoz/'
driver.get(base_url)

course_titles = []
study_options = []
start_dates = []
course_links = []
degree_types = []

# Find all rows containing course information
course_rows = driver.find_elements(By.XPATH, '//table[@class="m-table views-table m-table--courses"]/tbody/tr')
for row in course_rows:
    # Extract course title
    title_element = row.find_element(By.CLASS_NAME, 'm-table__cell--title')
    course_title = title_element.find_element(By.TAG_NAME, 'a').text.strip()  # Trim spaces
    course_titles.append(course_title)

    
    # Extract study options
    options_element = row.find_element(By.CLASS_NAME, 'm-table__cell--field_study_modes_field_study_mode_ref_name')
    study_option_links = options_element.find_elements(By.TAG_NAME, 'a')
    study_option_texts = [option.text for option in study_option_links]
    study_options.append(', '.join(study_option_texts))
    
        
#     # Extract Degree Type
#     degree_type = options_element.text.split(' - ')[-1].strip()
#     degree_types.append(degree_type)


    # Extract Degree Type
    degree_text = options_element.text
    if '-' in degree_text:
        degree = degree_text.split(' - ')[-1].strip() 
        if 'Full-time' in degree or 'Part-time' in degree:
            degree_type = 'NA'
        else:
            degree_type = degree
    else:
        degree_type = 'NA'
    degree_types.append(degree_type)
    
    # Extract start dates
    start_date_element = row.find_element(By.CLASS_NAME, 'm-table__cell--start_year_link')
    start_date_links = start_date_element.find_elements(By.TAG_NAME, 'a')
    start_date_texts = [date.text.strip() for date in start_date_links]  # Trim spaces
    start_dates.append(', '.join(start_date_texts))
    
    # Extract course links
    course_link = title_element.find_element(By.TAG_NAME, 'a').get_attribute('href')
    course_links.append(course_link)
    


# Create a DataFrame to store the extracted data
data = {
    'Course_Title': course_titles,
    'Degree_Type': degree_types,
    'Attendance_Mode': study_options,
    'Start_Date': start_dates,
    'Course_Link': course_links

}


df = pd.DataFrame(data)

# Replace empty or null values with 'NA'
df = df.fillna('NA')

# Trim leading and trailing spaces
df = df.applymap(lambda x: x.strip() if isinstance(x, str) else x)

df

driver.quit()
```
