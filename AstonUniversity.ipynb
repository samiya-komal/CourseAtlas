```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd

# Chrome webdriver
driver = webdriver.Chrome()

url = 'https://www.aston.ac.uk/courses-atoz'
driver.get(url)

# Find all course elements
course_elements = driver.find_elements(By.CSS_SELECTOR,'.views-row')

elements = driver.find_elements(By.XPATH, "//div[@class='views-row']//div[@class='title']")


course_names = []
degrees = []
dates = []
course_links=[]

for a in course_elements:

    element = a.find_element(By.CLASS_NAME, 'title')
    
    # Extracting the main title
    main_title = element.text.split('-')
    
    try:
        course_name = main_title[0].strip()
        course_names.append(course_name)
    except:
        course_names.append('NA')
        
    try:    
        degree_type_full = main_title[1].strip()
        degree_type = degree_type_full.split('(')[0].strip()
        degrees.append(degree_type)
    except:
        degrees.append('NA')
    

    try:
        date_element = element.find_element(By.XPATH, "./time")
        # date = date.get_attribute("datetime")
        date = date_element.text
        dates.append(date)
    except Exception as e:
        dates.append("NA")
        
    try:
        course_link = a.find_element(By.CSS_SELECTOR,'a').get_attribute('href')
        course_links.append(course_link)
    except:
        course_links.append('NA')  # Keep appending 'NA' if link extraction fails


# Create a dataframe
data = {
    "Course_Title": course_names,
    "Degree_Level": degrees,
    "Start_Date": dates,
    "Course_Links":course_links
}
df = pd.DataFrame(data)

driver.quit()
```
